```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
plt.rcParams.update({
    # Font
    "font.weight": "bold",
    "font.size":25,

    # Axes
    "axes.labelweight": "bold",
    "axes.titleweight": "bold",
    "axes.titlesize": 30,
    "axes.labelsize": 30,

    # Tick labels
    "xtick.labelsize": 25,
    "ytick.labelsize": 25,

    # Legend
    "legend.fontsize": 25,
    "legend.title_fontsize": 20,

    # Figure title
    "figure.titleweight": "bold",
    "figure.titlesize": 20,

    # Math text (matches regular font)
    "mathtext.default": "bf",
})
```

```python
df_real_temporal = pd.read_csv("../../data/project_39/epidemic_real_temporal.csv")
df_real_aggregate = pd.read_csv("../../data/project_39/epidemic_real_aggregate.csv")
df_ER_static = pd.read_csv("../../data/project_39/epidemic_ER_static.csv")
df_ER_temporal = pd.read_csv("../../data/project_39/epidemic_ER_temporal.csv")
```

```python
beta_values = df_real_temporal["beta"]

def my_plot(df, ax, color, markersize=10) :
    ax.errorbar(beta_values, df["avg_frac_inf"], yerr=df["std_frac_inf"], linestyle='None', 
    color=color, markersize=10,marker="o", label="I")
    #ax.errorbar(beta_values, df["avg_frac_recov"], yerr=df["std_frac_recov"], 
    #color=color,  markersize=markersize, label="R", marker="s", linestyle='none')

fig, ax = plt.subplots(figsize=(9, 6))

ax.set_xlabel(r"$ \beta $")
ax.set_ylabel("I/N")
ax.set_xticks([0.05, 0.25, 0.5, 1] )
ax.tick_params(axis="both", which="major", width=3)
#ax.set_yticks(np.round(np.linspace(0, 0.17, num=10), decimals=2))
my_plot(df_real_temporal, ax, "blue")
my_plot(df_real_aggregate, ax, "black")

system1_patch = mpatches.Patch(color='blue', label='real-temporal')
system2_patch = mpatches.Patch(color='black', label='real-aggregate')

ax.legend(handles=[system1_patch, system2_patch], fontsize=21, loc="center right")
ax.grid(1)

plt.savefig(fname= "../../data/project_39/figures/1.png", dpi=300,bbox_inches="tight")
```
